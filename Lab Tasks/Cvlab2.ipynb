```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt

image = cv2.imread(r"C:\Users\Hp\Downloads\1 (1).jpg" , cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(12, 8))
plt.subplot(3, 3, 1)
plt.title("Grayscale X-ray")
plt.imshow(image, cmap = 'gray')
plt.axis("off")

eq_image = cv2.equalizeHist(image)


plt.subplot(3, 3, 2)
plt.title("Equalized X-ray")
plt.imshow(eq_image, cmap = 'gray')
plt.axis("off")

colorMap = cv2.applyColorMap(eq_image, cv2.COLORMAP_JET)
plt.subplot(3, 3, 3)
plt.title("ColorMap X-ray")
plt.imshow(colorMap)
plt.axis("off")

adjusted = np.clip(colorMap * [0.2 , 0.4 , 0.6] , 0 , 255).astype(np.uint8)
plt.subplot(3, 3, 4)
plt.title("Color Adjusted X-ray")
plt.imshow(adjusted)
plt.axis("off")

rgb = cv2.cvtColor(adjusted,cv2.COLOR_BGR2RGB)
lowercolor = np.array([0,0,0])
uppercolor = np.array([230,240,250])
mask = cv2.inRange(adjusted, lowercolor, uppercolor)
filtered_image = cv2.bitwise_and(rgb, rgb, mask=mask)

plt.subplot(3, 3, 5)
plt.title("Filtered X-ray")
plt.imshow(filtered_image)
plt.axis("off")


c = 225 / np.log(1 + np.max(filtered_image))
log_transformed_image = c * np.log(1 + filtered_image)
log_transformed_image = np.array(log_transformed_image, dtype=np.uint8)

plt.subplot(3, 3, 6)
plt.title("log X-ray")
plt.imshow(log_transformed_image)
plt.axis("off")


gamma = 3.0
gamma_corrected_image = np.power(log_transformed_image, gamma)
gamma_corrected_image = np.clip(gamma_corrected_image * (255 / np.max(gamma_corrected_image)), 0, 255).astype(np.uint8)
plt.subplot(3, 3, 7)
plt.title('Power-Law Transformation')
plt.imshow(gamma_corrected_image)
plt.axis('off')
```

```python
import cv2
import numpy as np

video_path = r"C:\Users\Hp\Desktop\WhatsApp Video 2024-09-16 at 11.02.46_e005d114.mp4"
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

def histogram_equalization(frame):
    ycrcb = cv2.cvtColor(frame, cv2.COLOR_BGR2YCrCb)
    channels = cv2.split(ycrcb)
    cv2.equalizeHist(channels[0], channels[0])
    ycrcb = cv2.merge(channels)
    return cv2.cvtColor(ycrcb, cv2.COLOR_YCrCb2BGR)

def apply_heatmap(frame):
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    heatmap = cv2.applyColorMap(gray, cv2.COLORMAP_JET)
    return heatmap

def color_balance(frame):
    result = cv2.convertScaleAbs(frame, alpha=1.2, beta=20)
    return result

def log_transform(frame):
    frame = frame.astype(np.float32)
    c = 255 / np.log(1 + np.max(frame))
    log_image = c * (np.log(frame + 1e-5))
    log_image = np.clip(log_image, 0, 255)
    return log_image.astype(np.uint8)

def power_law_transform(frame, gamma=0.5):
    gamma_corrected = np.array(255 * (frame / 255) ** gamma, dtype='uint8')
    return gamma_corrected

frame_width = 640
frame_height = 360

while cap.isOpened():
    ret, frame = cap.read()
    
    if not ret:
        print("End of video or error encountered.")
        break

    frame = cv2.resize(frame, (frame_width, frame_height))
    
    hist_eq_frame = histogram_equalization(frame)
    heatmap_frame = apply_heatmap(frame)
    color_balanced_frame = color_balance(frame)
    log_transformed_frame = log_transform(frame)
    gamma_corrected_frame = power_law_transform(frame)

    hist_eq_frame = cv2.resize(hist_eq_frame, (frame_width, frame_height))
    heatmap_frame = cv2.resize(heatmap_frame, (frame_width, frame_height))
    color_balanced_frame = cv2.resize(color_balanced_frame, (frame_width, frame_height))
    log_transformed_frame = cv2.resize(log_transformed_frame, (frame_width, frame_height))
    gamma_corrected_frame = cv2.resize(gamma_corrected_frame, (frame_width, frame_height))

    top_row = np.hstack((frame, hist_eq_frame))
    bottom_row = np.hstack((heatmap_frame, color_balanced_frame, log_transformed_frame, gamma_corrected_frame))

    top_row = cv2.resize(top_row, (frame_width * 2, frame_height))
    bottom_row = cv2.resize(bottom_row, (frame_width * 2, frame_height))

    combined_frame = np.vstack((top_row, bottom_row))

    cv2.imshow('Video Processing Comparison', combined_frame)

    if cv2.waitKey(30) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
